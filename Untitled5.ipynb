```python
plt.figure(figsize=(100,100))
sns.countplot(x='Vict Age', data=df)
plt.title('Victim Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Vict Sex', data=df)
plt.title('Victim Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Vict Descent', data=df)
plt.title('Victim Descent Distribution')
plt.xlabel('Descent')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Weapon Desc', data=df)
plt.title('Weapon Description Distribution')
plt.xlabel('Weapon Description')
plt.ylabel('Count')
plt.show()
```

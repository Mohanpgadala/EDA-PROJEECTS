```python
# @title Average Movie Rating Over Time by Genre

import matplotlib.pyplot as plt

# Group the data by year and genre, and calculate the average rating
avg_ratings = df.groupby(['Year', 'Genre'])['IMDb Rating'].mean().unstack()

# Plot the line chart
avg_ratings.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average IMDb Rating')
_ = plt.title('Average Movie Rating Over Time by Genre')
```

```python
# @title Distribution of Movie Ratings

import matplotlib.pyplot as plt

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['IMDb Rating'], bins=10, edgecolor='black')
plt.xlabel('IMDb Rating')
plt.ylabel('Frequency')
_ = plt.title('Distribution of Movie Ratings')
```

```python
# @title Average Movie Rating by Certificate

import matplotlib.pyplot as plt

# Group the data by certificate and calculate the average rating
ratings_by_certificate = df.groupby('Certificates')['IMDb Rating'].mean()

# Plot the bar chart
ratings_by_certificate.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Certificate')
plt.ylabel('Average IMDb Rating')
_ = plt.title('Average Movie Rating by Certificate')
```
